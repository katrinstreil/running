```python
import datetime

def plot_halfmarathon(goal_time, heartrate_range, ax):
    max_hr = collated_compact.max_heartrate()
    pace = goal_time / 21.2  # Sekunden pro km

    hr_min_rel, hr_max_rel = heartrate_range
    hr_avg = (hr_min_rel + hr_max_rel) / 2 * max_hr
    hr_err_low = (hr_avg - hr_min_rel * max_hr)
    hr_err_high = (hr_max_rel * max_hr - hr_avg)

    ax.errorbar(
        x=[pace],
        y=[hr_avg],
        yerr=[[hr_err_low], [hr_err_high]],
        fmt='',
        color='tab:red'
    )
    
    # Konvertiere goal_time (Sekunden) in hh:mm:ss
    formatted_time = str(datetime.timedelta(seconds=int(goal_time)))
    
    ax.text(
        pace * 0.95,
        hr_avg - hr_err_low * 3,
        f"HM in\n{formatted_time}",
        color='tab:red'
    )


def plot_marathon(goal_time, heartrate_range, ax):
    max_hr = collated_compact.max_heartrate()
    pace = goal_time / 42.2  # Sekunden pro km

    hr_min_rel, hr_max_rel = heartrate_range
    hr_avg = (hr_min_rel + hr_max_rel) / 2 * max_hr
    hr_err_low = (hr_avg - hr_min_rel * max_hr)
    hr_err_high = (hr_max_rel * max_hr - hr_avg)

    ax.errorbar(
        x=[pace],
        y=[hr_avg],
        yerr=[[hr_err_low], [hr_err_high]],
        fmt='',
        color='darkred'
    )
    
    # Konvertiere goal_time (Sekunden) in hh:mm:ss
    formatted_time = str(datetime.timedelta(seconds=int(goal_time)))
    
    ax.text(
        pace * 0.95,
        hr_avg - hr_err_low * 3,
        f"M in\n{formatted_time}",
        color='darkred'
    )
```

```python
from collections import defaultdict
import os
import glob
import yaml
import re
import numpy as np
from mycolorpy import colorlist as mcp
import matplotlib.pyplot as plt


class CompactDataProcessor():
    def __init__(self, entry):
        self.date = entry["date"]
        self.distance = entry["distance_km"]  # in Meter
        self.average_heartrate = entry["heartrate_avg"]
        self.max_heartrate = entry["max_heartrate"]
        self.time = float(entry["time"])  # Sekunden
        self.pace = float(entry["pace"])  # Sekunden
        self.df = None
        self.cycle_day = int(entry["cycle_day"])  
        self.vo2max = float(entry["vo2max"])  
        
class CollatedCompactProcessors:
    def __init__(self, folder="data_compact"):
        self.folder = folder
        self.processors = self.load_all_from_compact_yaml()
        self.sorted_files, self.sorted_dates = self._get_fit_file_names_and_dates()
    
    def _get_fit_file_names_and_dates(self):
        """Returns sorted lists of FIT file names and their corresponding dates."""
        files = glob.glob(os.path.join(self.folder, "*.yaml"))
        
        file_dates =  [re.search("([0-9]{4}\-[0-9]{2}\-[0-9]{2})", fileName).group(1) for fileName in files]
        
        # Sort files by date
        sorted_files_dates = sorted(zip(files, file_dates), key=lambda x: x[1])
        sorted_files, sorted_dates = zip(*sorted_files_dates) if sorted_files_dates else ([], [])
        
        return list(sorted_files), np.array(sorted_dates)
        
    def load_all_from_compact_yaml(self):
        yaml_files = sorted(glob.glob(os.path.join(self.folder, "*.yaml")))
        processors = []
        for path in yaml_files:
            with open(path) as f:
                data = yaml.safe_load(f)
                runs = data.get("runs", [])
                for e in runs:
                    p = CompactDataProcessor(e)
                    processors.append(p)
        return processors

    def max_heartrate(self):
        _heart = [_.max_heartrate for _ in self.processors]
        return np.max(_heart)
    
    def _format_pace(self, pace_in_minutes):
        """Convert pace from minutes per kilometer to 'min' 'sec' format."""
        minutes = int(pace_in_minutes)
        seconds = int((pace_in_minutes - minutes) * 60)
        return f"{minutes}' {seconds}\""

    def get_color_for_distance(self):
        c_short = 'tab:orange'
        c_long = 'tab:green'
        c_5 = 'tab:red'
        c_sonst = 'tab:blue'
        colors, markers = [], []
        for p in self.processors:
            c = c_sonst
            if p.distance > 10000:
                c = c_long
            if p.distance < 2500:
                c = c_short
            if p.distance < 5100 and p.distance > 4900:
                c = c_5
            colors.append(c)
            if '2024-12' in p.date:
                markers.append('x')
            if '2025-01' in p.date:
                markers.append('o')
            if '2025-02' in p.date:
                markers.append('+')
            if '2025-03' in p.date:
                markers.append('*')
            if '2025-04' in p.date:
                markers.append('s')
            if '2025-05' in p.date:
                markers.append('.')
            if '2025-06' in p.date:
                markers.append('x')
            if '2025-07' in p.date:
                markers.append('+')
        return colors, markers
    
    
    
    def get_color_for_distance_month(self):
        c_12 = 'gold'
        c_01 = 'gold'
        c_02 = 'lightsalmon'
        c_03 = 'violet'
        c_04 = 'darkviolet'
        c_05 = 'indigo'
        c_06 = 'black'
        c_07 = 'black'
        
        
        
        p_12, p_01, p_02, p_03, p_04, p_05, p_06, p_07 = [],[],[],[],[],[], [], []
        #c_12, c_01, c_02, c_03, c_04, c_05= colors=mcp.gen_color(cmap="spring_r",n = 6)
        
        colors, markers = [], []
        for p in self.processors:
            if '2024-12' in p.date:
                colors.append(c_12)
                p_12.append(p)
            if '2025-01' in p.date:
                colors.append(c_01)
                p_01.append(p)
            if '2025-02' in p.date:
                colors.append(c_02)
                p_02.append(p)
            if '2025-03' in p.date:
                colors.append(c_03)
                p_03.append(p)
            if '2025-04' in p.date:
                colors.append(c_04)
                p_04.append(p)
            if '2025-05' in p.date:
                colors.append(c_05)
                p_05.append(p)
            if '2025-06' in p.date:
                colors.append(c_06)
                p_06.append(p)
                markers.append('x')  
            if '2025-07' in p.date:
                colors.append(c_07)
                p_07.append(p)
                markers.append('x')  
            else:           
                markers.append('o')  
        return colors, markers, p_12, p_01, p_02, p_03, p_04, p_05 , p_06, p_07,  c_12, c_01, c_02, c_03, c_04, c_05, c_06, c_07
    
    def plot_heart_rate_vs_pace(self, interval=1000, fit = True,
                               style = 'default'):
        """Plots all heart rate vs pace data from different files with a gradient color scale based on time."""
        if not self.processors:
            print("No data available for plotting.")
            return
        if style == 'default':
            colors=mcp.gen_color(cmap="spring_r",n = len(self.processors))
            markers = ['x','o','+','s','^','>', '*'] * 50
        elif style == 'month':
            colors, markers , p_12, p_01, p_02, p_03, p_04, p_05 , p_06, p_07, c_12, c_01, c_02, c_03, c_04, c_05, c_06, c_07= self.get_color_for_distance_month()
        else:
            colors, markers = self.get_color_for_distance()
        plt.figure(figsize=(8, 5))
        
        i = 0 
        paces, heartrates = [], []
        for processor in (self.processors):
            paces.append(processor.pace)
            heartrates.append(processor.average_heartrate)
            formatted_pace = self._format_pace(processor.pace)
            label_text = f"{processor.date} ({formatted_pace}, {(processor.distance/1000):.3} km)"
            plt.errorbar(processor.pace, processor.average_heartrate,  color=colors[i],  label=label_text, elinewidth = 0.2, markersize = markersize[markers[i]],marker = markers[i])
            i+=1
        
        if style == 'month':
            ps = [p_12, p_01, p_02, p_03, p_04, p_05 ]
            colors_= [c_12, c_01, c_02, c_03, c_04, c_05]
            for i,p in enumerate(ps):
                paces_mean = np.mean([_.pace for _ in p])
                paces_std = np.std([_.pace for _ in p])
                
                hr_mean = np.mean([_.average_heartrate for _ in p])
                hr_std = np.std([_.average_heartrate for _ in p])
                print(i)
                plt.errorbar(x = paces_mean, xerr = paces_std,
                             y = hr_mean, yerr = hr_std, color=colors_[i],  
                             elinewidth = 1.8, 
                             marker = '+')
                
            
            
        plt.xlabel("Pace (min/km)")
        plt.ylabel("Heart Rate (bpm)")
        plt.title("Heart Rate as a Function of Pace (Color by Time)")
        #plt.grid(True)
        plt.legend(loc="upper left", bbox_to_anchor=(1, 1), ncol = 2, fontsize="small",)
        
        
        if fit: 
            #m, b = np.polyfit(paces, heartrates, 2)
            pace_range = np.linspace(min(paces), max(paces), 1000)
            #fit_line = m * pace_range + b
            z = np.polyfit(paces, heartrates, 3)
            p = np.poly1d(z)
            _ = plt.plot(pace_range, p(pace_range), '-')
            fit_line  = p(pace_range)
            #plt.plot(pace_range, fit_line, color="grey",
            #         linestyle="--", alpha= 0.3)#, label=f"Fit: HR = {m:.2f} * Pace + {b:.2f}")
            distances = []
            for p_x, p_y in zip(paces, heartrates):
                ddx = (pace_range - p_x)
                ddy = (fit_line - p_y)
                dd = np.min(np.sqrt(ddx**2 + ddy**2))
                
                sign = -1
                pace_soll = pace_range[np.argmin(( pace_range- p_x)**2)]
                heartrate_soll =  p(pace_soll)# m * pace_soll + b
                if heartrate_soll > p_y:
                    sign = +1 
                distances.append(dd *sign )
            self.distances = distances
        #plt.show()
        return plt.gca()

    def plot_distances_vs_cycle_day(self):
        
        for pro , dd in zip(self.processors, self.distances):
            plt.scatter(pro.cycle_day, dd, color= 'tab:blue')
            #plt.scatter(pro.pace(), dist, color= 'tab:blue')
        plt.scatter(self.processors[-1].cycle_day, self.distances[-1], marker = 'x')
        plt.ylabel("Deviation from Fit [arb. unit]")
        plt.xlabel("Cycle Day")
        plt.gca().set_ylim(np.min(self.distances) -2, np.max(self.distances) + 2)
        yy = plt.gca().get_ylim()
        
        if True:
            plt.fill_betweenx(y = (yy[0],yy[1]), x1 = 0, x2 = 4,  color = 'tab:red', alpha = 0.2)
            plt.text(-1, yy[1]- 1 , 'Menstruation', color = 'tab:red')
            plt.fill_betweenx(y = (yy[0],yy[1]), x1 = 4, x2 = 13,  color = 'tab:orange', alpha = 0.2)
            plt.text(6.6, yy[1]- 1 , 'Follicular', color = 'tab:orange')
            plt.fill_betweenx(y = (yy[0],yy[1]), x1 = 13, x2 = 15,  color = 'tab:green', alpha = 0.2)
            plt.text(12, yy[1]- 1 , 'Ovulation', color = 'tab:green')
            plt.fill_betweenx(y = (yy[0],yy[1]), x1 = 15, x2 = 28,  color = 'tab:purple', alpha = 0.2)
            plt.text(20, yy[1]- 1 , 'Luteal', color = 'tab:purple')
            
        plt.plot([0,28], [0,0], color = 'grey', alpha = 0.4)
        plt.text(25, 2, 'better',  color = 'grey',)       
        plt.text(25, -2, 'worse',  color = 'grey',)
    
        
    def plot_distances_vs_date(self):
        
        for i, dd in enumerate( self.distances):
            plt.scatter(i, dd, color= 'tab:blue')
            #plt.scatter(pro.pace(), dist, color= 'tab:blue')
        plt.scatter(self.processors[-1].cycle_day, self.distances[-1], marker = 'x')
        plt.ylabel("Deviation from Fit [arb. unit]")
        plt.xlabel("Training Day")
        plt.gca().set_ylim(np.min(self.distances) -2, np.max(self.distances) + 2)
        yy = plt.gca().get_ylim()
        
        plt.plot([0,28], [0,0], color = 'grey', alpha = 0.4)
        plt.text(25, 2, 'better',  color = 'grey',)       
        plt.text(25, -2, 'worse',  color = 'grey',)
    
    def plot_monthly_pace_trends(self, bin_size=1):
        import matplotlib.pyplot as plt
        from collections import defaultdict

        # Monatsdaten und Farben
        _, _, p_12, p_01, p_02, p_03, p_04, p_05, p_06, p_07, \
            c_12, c_01, c_02, c_03, c_04, c_05, c_06, c_07 = self.get_color_for_distance_month()

        month_data = {
            "2024-12": (p_12, c_12),
            "2025-01": (p_01, c_01),
            "2025-02": (p_02, c_02),
            "2025-03": (p_03, c_03),
            "2025-04": (p_04, c_04),
            "2025-05": (p_05, c_05),
            "2025-06/07": (p_06 + p_07, c_06),  # ðŸ‘ˆ kombiniert!
        }

        # Pace-Grenzen
        all_paces = [p.pace for p in self.processors if p.pace is not None]
        pace_min = int(min(all_paces))
        pace_max = int(max(all_paces)) + 1

        # Vertikale Binlinien (1-Minuten-Breite)
        for p in range(pace_min, pace_max + bin_size, bin_size):
            plt.axvline(x=p, color='lightgrey', linestyle='--', linewidth=0.5)

        # Linien pro Monat mit Fehlerband
        for month, (runs, color) in month_data.items():
            if not runs:
                continue

            bin_dict = defaultdict(list)
            for r in runs:
                if r.pace is None or r.average_heartrate is None:
                    continue
                pace_bin = int(r.pace // bin_size) * bin_size
                bin_dict[pace_bin].append(r)

            bin_centers = []
            avg_hrs = []
            std_hrs = []

            for pace_bin in sorted(bin_dict.keys()):
                bin_runs = bin_dict[pace_bin]
                if not bin_runs:
                    continue
                avg_pace = np.mean([r.pace for r in bin_runs])
                avg_hr = np.mean([r.average_heartrate for r in bin_runs])
                std_hr = np.std([r.average_heartrate for r in bin_runs])

                bin_centers.append(avg_pace)
                avg_hrs.append(avg_hr)
                std_hrs.append(std_hr)

            if len(bin_centers) >= 2:
                plt.plot(bin_centers, avg_hrs, color=color, label=month, marker = '.')
                plt.fill_between(bin_centers,
                                 np.array(avg_hrs) - np.array(std_hrs),
                                 np.array(avg_hrs) + np.array(std_hrs),
                                 color=color,
                                 alpha=0.2)

        # Labels & Legende
        plt.xlabel("Pace (min/km)")
        plt.ylabel("Heart Rate (bpm)")
        plt.title("Monthly Pace Trends with HR Error Bands")
        plt.legend(loc="upper left", bbox_to_anchor=(1, 1), ncol=2, fontsize="small")

        
    def analyze_pace_change_at_fixed_hr(self, target_hr=160):
        from sklearn.linear_model import LinearRegression
        import numpy as np

        X, y = [], []

        for p in self.processors:
            if p.pace is not None and p.average_heartrate is not None:
                # Monat als Zahl extrahieren
                match = re.search(r"(\d{4})-(\d{2})", p.date)
                if not match:
                    continue
                year, month = int(match.group(1)), int(match.group(2))
                month_index = (year - 2024) * 12 + month  # z.â€¯B. Jan 2024 = 1, Feb = 2, ...

                X.append([month_index, p.average_heartrate])
                y.append(p.pace)

        if len(X) < 5:
            print("Nicht genug Daten fÃ¼r Regression.")
            return

        X = np.array(X)
        y = np.array(y)

        model = LinearRegression()
        model.fit(X, y)

        month_coef = model.coef_[0]
        hr_coef = model.coef_[1]
        intercept = model.intercept_

        print("ðŸ“ˆ Regressionsergebnisse:")
        print(f"  Steigung pro Monat:         {month_coef:.3f} min/km")
        print(f"  Steigung pro bpm HF:        {hr_coef:.3f} min/km")
        print(f"  Achsenabschnitt (Offset):   {intercept:.3f} min/km")

        predicted_pace = model.predict([[X[-1,0], target_hr]])[0]
        print(f"\nðŸ” Prognose: Pace bei {target_hr} bpm im aktuellen Monat: {predicted_pace:.2f} min/km")

        return {
            "pace_per_month": month_coef,
            "pace_per_bpm": hr_coef,
            "intercept": intercept,
            "predicted_pace": predicted_pace
        }
```
